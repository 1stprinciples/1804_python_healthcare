# An introduction to matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
# The following line helps charts to be displayed in Jupyter Notebooks
%matplotlib inline
```

## A simple line chart

```python
x = np.arange(1,10.01,0.01) # array from 1 to 10 in steps of 0.01
y = np.sin(x)
plt.plot(x,y)
plt.xlabel ('x')
plt.ylabel ('sin(x)')
plt.title('A sine wave')
plt.show()
```

## Two lines with legend

```python
x = np.arange(1,10.01,0.01) # array from 1 to 10 in steps of 0.01
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label='sin')
plt.plot(x,y2,label='cos')
plt.legend()
plt.xlabel ('x')

# If you want to save a plot, then save before showing
plt.savefig('line_plot.png')
            
plt.show()
```

## Customizing a line plot

```python
x = np.arange(1,10.01,0.01) # array from 1 to 10 in steps of 0.01
y = np.sin(x)
plt.plot(x,y,
         linestyle='--',
         color='g',
         marker='o',
         markeredgecolor='k',
         markerfacecolor='w',
         markersize=10,
         markevery=50)

# Adding options (keywaord arguments. Note the dastardly American spelling)
# Use help (plt.plot to see all options)
plt.xlabel ('x')
plt.ylabel ('sin(x)')
plt.show()
```

## A scatter plot

```python
x = np.random.rand(250) * 100 # 250 values between 0 and 100
jitter = 10 - np.random.rand(250) * 250 # 100 values between -10 and 10
y = x + jitter

# Scatter plot with transparency of points (alpha)
plt.scatter(x, y,
            color='r',
            alpha = 0.5)

plt.show()
```

## A bar chart

```python
x = [1, 2, 3, 4]
y = [5.,25.,50.,20.]
 
labels = ['A', 'B', 'C', 'D']
plt.bar(x, y)
# To set xlabels we need to pass location (an array starting at 1, and labels)
plt.xticks(np.arange(1,5), labels)
plt.show()
```

## Boxplots

Box plots display a box for the interquartile range, median as organe line, range without outliers, and outliers.

Outliers usually defined as 1.5*IQR (interquartile range) outside IQR.

```python
x=np.random.randn(1000,4) # samples from a normal distribution

labels = ['A', 'B', 'C', 'D']
plt.boxplot(x)
# To set xlabels we need to pass location (an array starting at 1, and labels)
plt.xticks(np.arange(1,5), labels)

plt.show()
```

To avoid plotting outliers separately change the `whis` (whiskers) argument from default 1.5 to a high value.

```python
x=np.random.randn(1000,4) # samples from a normal distribution

labels = ['A', 'B', 'C', 'D']
plt.boxplot(x, whis=1000)
# To set xlabels we need to pass location (an array starting at 1, and labels)
plt.xticks(np.arange(1,5), labels)

plt.show()
```
